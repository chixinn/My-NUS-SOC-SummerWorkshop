```python
import re
import requests
```

### Single Matching

```python
html = "<html><body>Web Mining<br/>Web mining is concerned with mining data from the web the web which can then be transformed into knowledge</body></html>"
pattern = "^<html><body>(.*)<br/>(.*)</body></html>$"
```

```python
#group 1
title = re.search(pattern, html).group(1)
title
```

```python
#group 2
description = re.search(pattern, html).group(2)
description
```

```python
#match whole pattern
match = re.search(pattern, html).group()
match
```

### Multiple Matching

```python
html2 = "<html><body> <ul><li>johndoe@gmail.com</li><li>janedoe@gmail.com</li><li>Robin</li></ul></body></html>"
```

```python
pattern2 = "<li>(.*?)</li>"
```

```python
list_values = re.findall(pattern2, html2)
list_values
```

### Webscraping using regular expression on real website

```python
page = requests.get("https://sws.comp.nus.edu.sg/")
page.content
```

```python
print(type(page.content))

#convert from bytes to str
html3 = page.content.decode("utf-8")
html3
```

```python
#. by default does not match for newline characters
#re.DOTALL - will make . match even for newline
pattern3 = '<div id="WEcc664a539a".*?>.*?<span.*?>(.*?)</span>.*?</div>'
results = re.search(pattern3, html3, re.DOTALL)
results.group()
```

```python
results.group(1)
```

```python
extracted_text = results.group(1)
print("Original text:")
print(extracted_text)
print("=============")

#remove some html
#\\1 - refers to group 1
extracted_text = re.sub("<br />", "\n", extracted_text)
extracted_text = re.sub("&nbsp;", " ", extracted_text)
extracted_text = re.sub("<.*?>(.*?)</.*>", "\\1", extracted_text)

print("Processed text:")
print(extracted_text)
```
