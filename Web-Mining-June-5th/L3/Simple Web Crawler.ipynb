```python
import requests
from bs4 import BeautifulSoup
from bs4.element import Comment
import re
from pandas import DataFrame
import time
```

```python
STARTING_URL = "https://sws.comp.nus.edu.sg/"
```

```python
urls = []
htmls = []
texts = []
urls.append(STARTING_URL)
page = requests.get(STARTING_URL)
```

```python
#get the raw HTML
htmls.append(page.content)
page.content
```

```python
#parse the HTML
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
#pretty print HTML
print(soup.prettify())
```

```python
text = soup.text.strip()
texts.append(text)
text
```

```python
#find all links
links = soup.find_all('a')
links
```

```python
#get href attribute of all links
links_href = []
for link in links:
    href = link.get('href')
    if not href.startswith("http"):
        href = STARTING_URL + href
        links_href.append(href)        

def filter_link(link):
    return link.endswith('htm') or link.endswith('html')

links_href = list(filter(filter_link, links_href))
links_href = list(set(links_href))
links_href
```

```python
for url in links_href:
    time.sleep(0.1)
    page = requests.get(url)
    html = page.content
    htmls.append(html)
    soup = BeautifulSoup(page.content, 'html.parser')
    text = soup.text.strip()
    urls.append(url)
    texts.append(text)
    
result = DataFrame({'url': urls, 'text': texts, 'html': htmls})
result
```
